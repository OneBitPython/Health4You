```python
from tensorflow import keras
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
train_data
```

```python
train_data = train_data.dropna()
train_data.isna().sum()
```

```python
convert_gender = LabelEncoder().fit(train_data['Gender'])
train_data['binary_gender'] = convert_gender.transform(train_data['Gender'])
```

```python
convert_work_from_home = LabelEncoder().fit(train_data['WFH Setup Available'])
train_data['binary_work_from_home'] = convert_work_from_home.transform(train_data['WFH Setup Available'])
```

```python
convert_company_type = LabelEncoder().fit(train_data['Company Type'])
train_data['binary_company_type'] = convert_company_type.transform(train_data['Company Type'])
```

```python
train_data
```

```python
features = train_data[['binary_gender', "binary_work_from_home", "Designation", "Resource Allocation", "Mental Fatigue Score"]]
target = train_data['Burn Rate']
```

```python
company_type_classes = ['Product', "Service"]
gender_classes = ['Female', 'Male']
work_from_home_classes = ['No',"Yes"]
```

```python
train_data.head(5)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features, target, train_size=0.95)
```

```python
from catboost import CatBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor

model = CatBoostRegressor().fit(x_train, y_train)
model.save_model("burnout_model")
```

```python
test_data['binary_gender'] = convert_gender.transform(test_data['Gender'])
test_data['binary_work_from_home'] = convert_work_from_home.transform(test_data['WFH Setup Available'])
test_data['binary_company_type'] = convert_company_type.transform(test_data['Company Type'])
```
